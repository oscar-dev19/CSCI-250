```
import numpy as np

mmatrix= np.array([[1,2,3],[4,5,6]])
print(mmatrix)
```

```
#Sigmoid
#sigmoid function
def sigmoid(X):
   return 1/(1+np.exp(-X))

#Example with mmatrix defined above
sigmoid(mmatrix)

```

```
#ReLu function
def relu(X):
   return np.maximum(0,X)
#Example with mmatrix defined above
relu(mmatrix)
```

```
#softmax function
def softmax(X):
    expo = np.exp(X)
    expo_sum = np.sum(np.exp(X))
    return expo/expo_sum
#Example with mmatrix defined above
print (softmax(mmatrix))
```
