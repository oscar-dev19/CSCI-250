```python
"""
Simple program that graphs 'happiness' in 
relation to world gdp per capita data.

Author: Oscar Lopez
"""
import pandas as pd
import matplotlib.pyplot as plt

#Setting up column names.
col_names = ['Country', 'Happiness_Rank', 'Happiness_Score', 'Whisker_High', 'Whisker_Low', 'GDP_Per_Capita', 'Family', 'Health_Life_Expectancy', 'Freedom', 'Genorisity', 'Trust_Government_Corruption', 'Dystopia_Residual']

#Loading data
data_url = "https://raw.githubusercontent.com/ebubekiraygun/World-Happiness-Data-on-Python/master/2017.csv"
pima = pd.read_csv(data_url, header=None, names=col_names, skiprows=1)

#Setting up data for gdp vs happiness graph and plotting data.
plt.plot(pima.GDP_Per_Capita, pima.Happiness_Score, 'o')
plt.title('Oscar Lopez - GDP_Per_Capita vs. Happiness_Score')
plt.legend(['GDP_Per_Capita'])
plt.show()
```

```python
#Setting up data for freedom vs happiness graph and plotting data.
plt.plot(pima.Freedom, pima.Happiness_Score, 'o', color='green')
plt.title('Oscar Lopez - Freedom vs Happiness_Score')
plt.legend(['Freedom'])
plt.show()
```

```python
#Creating heatmap correlation
corr = pima.corr()
corr
```

```python
#Creating visual for heatmap correlation graph.
import seaborn as sb

sb.heatmap(corr, annot = True)
```
